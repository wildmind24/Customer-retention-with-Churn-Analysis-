```python
#Cleaning Data
#Import libraries and load data frame 

import pandas as pd

df = pd.read_csv(r"C:\Users\USER\Desktop\PAUL'S PORTFOLIO\churn\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
# Display basic info  
print(df.info())  

# Display first 5 rows  
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
#from basic info, TotalCharges was saved as an object 
```

```python
# Convert 'TotalCharges' to numeric, forcing errors to NaN
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')

# Check if there are any NaN values after conversion
print(df["TotalCharges"].isna().sum())
```

```python
#Check rows where TotalCharges is NaN
df[df["TotalCharges"].isna()]
```

```python
# Fill NaN TotalCharges with 0
df["TotalCharges"].fillna(0, inplace=True)

# Verify that there are no more missing values
print(df["TotalCharges"].isna().sum())
```

```python
# Summary statistics for numeric columns
print(df.describe())

# Distribution of categorical columns (e.g., gender, partner, etc.)
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\n{col} distribution:")
    print(df[col].value_counts())

# Correlation between numeric features (tenure, MonthlyCharges, TotalCharges)
correlation = df[['tenure', 'MonthlyCharges', 'TotalCharges']].corr()
print("\nCorrelation matrix:")
print(correlation)
```

```python
#Data visualization 
```

```python
#Import libraries 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pairplot for numerical features
sns.pairplot(df[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']], hue='Churn', palette='Set2')
plt.suptitle('Pairplot of Numerical Features vs Churn', y=1.02)
plt.show()
```

```python
# Stacked bar plot for Churn vs Contract
contract_churn = pd.crosstab(df['Contract'], df['Churn'])
contract_churn.plot(kind='bar', stacked=True, color=['red', 'green'], figsize=(8, 5))
plt.title('Churn vs Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Stacked bar plot for Churn vs Payment Method
paymentmethod_churn = pd.crosstab(df['PaymentMethod'], df['Churn'])
paymentmethod_churn.plot(kind='bar', stacked=True, color=['red', 'green'], figsize=(8, 5))
plt.title('Churn vs Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Stacked bar plot for Churn vs Internet Service
internetservice_churn = pd.crosstab(df['InternetService'], df['Churn'])
internetservice_churn.plot(kind='bar', stacked=True, color=['red', 'green'], figsize=(8, 5))
plt.title('Churn vs Internet Service')
plt.xlabel('Internet Service')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Violin plot for MonthlyCharges by Churn
plt.figure(figsize=(8, 5))
sns.violinplot(x='Churn', y='MonthlyCharges', data=df, palette='Set2')
plt.title('Distribution of Monthly Charges by Churn')
plt.show()

# Violin plot for TotalCharges by Churn
plt.figure(figsize=(8, 5))
sns.violinplot(x='Churn', y='TotalCharges', data=df, palette='Set2')
plt.title('Distribution of Total Charges by Churn')
plt.show()
```

```python
# Full correlation matrix visualization
correlation_matrix_full = df.select_dtypes(include=['float64', 'int64']).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_full, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Full Dataset)')
plt.show()
```

```python
# Boxplot for Tenure vs Churn
plt.figure(figsize=(8, 5))
sns.boxplot(x='Churn', y='tenure', data=df, palette='Set2')
plt.title('Tenure vs Churn')
plt.show()
```
